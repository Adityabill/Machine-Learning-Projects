```python
#importing libraries
import numpy as num
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
```

```python
#importing and inspection of dataset
dataset = pd.read_csv("/kaggle/input/advtlr/Advertising.csv")
print(dataset.shape)
print(dataset.head(10))
dataset.describe()
```

```python
#Scatter plot between variables
dataset.plot(x="TV", y="Sales", style="o")
plt.title("TV Spend v/s Sales")
plt.xlabel("TV Spends")
plt.ylabel("Sales")
plt.show()
```

```python
#Setting up input and output variables
#Selecting 2nd, 3rd and 4th columns
X = dataset.iloc[:, 1:4]
#selecting 4th column
Y = dataset.iloc[:, 4]
```

```python
#Fitting of linear regression model
regressor = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

regressor.fit(X_train, Y_train)
```

```python
#inspection of model

#To retrieve the intercepts->
print(regressor.intercept_)
#To retrieve the slop->
print(regressor.coef_)
```

```python
#Prediction on test sets
Y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': Y_test.values, 'Predicted': Y_pred})
ax1 = df.plot.scatter(x='Actual', y='Predicted')
Y_pred.shape
```

```python
#Checking the validity of the model
X2 = sm.add_constant(X_train)
est = sm.OLS(Y_train, X2)
est2 = est.fit()
print(est2.summary())
```
